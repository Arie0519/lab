```python
# ライブラリのインポート
import pandas as pd
import glob
import statistics
```

```python
# データが保存されているフォルダのパスを指定
folder_path = R'C:/Users/stomo/Desktop/study/datasets/estdistdata/'
```

```python
file_pattern = '*.csv'
files = glob.glob(f'{folder_path}/{file_pattern}')
print(files)
```

```python
# ファイルを数字ごとにグループ化
data_frames = {}
for file in files:
    df = pd.read_csv(file)
    df = df.dropna(how="any")
    df['timeStamp'] = df['timeStamp'] / 1000000000
    
    # ファイル名から数字部分を抽出
    parts = file.split('_')
    for part in parts:
        if part.isdigit():
            number = int(part)
            if number not in data_frames:
                data_frames[number] = []
            data_frames[number].append(df)
            break
print(data_frames)
```

```python
# 正解距離とアプリで予測された距離を表示
dists = []
for number, dfs in data_frames.items():
    dist_list = set()
    for df in dfs:
        dist_list.add(df['distance'].iloc[-1])
    dist = statistics.mean(dist_list)
    dists.append([number/100, dist])
columns = ["distance", "predicted"]
dists = pd.DataFrame(dists, columns=columns)
dists = dists.sort_values("distance")
dists = dists.reset_index(drop=True)
print(dists)

# csvファイルに保存
dists.to_csv(R'C:/Users/stomo/Desktop/study/logs/distance.csv', index=False)
```
